```python
from collections import deque, defaultdict

graph_50 = {
    'G_F': ['G_0', 'G_1', 'G_2'],
    'G_0': ['G_F', 'G_3', 'G_4', 'G_5'],
    'G_1': ['G_F', 'G_6', 'G_7', 'G_8'],
    'G_2': ['G_F', 'G_9', 'G_10', 'G_11'],
    'G_3': ['G_0', 'G_12', 'G_13'],
    'G_4': ['G_0', 'G_14', 'G_15'],
    'G_5': ['G_0', 'G_16', 'G_17'],
    'G_6': ['G_1', 'G_18', 'G_19'],
    'G_7': ['G_1', 'G_20', 'G_21'],
    'G_8': ['G_1', 'G_22', 'G_23'],
    'G_9': ['G_2', 'G_24', 'G_25'],
    'G_10': ['G_2', 'G_26', 'G_27'],
    'G_11': ['G_2', 'G_28', 'G_29'],
    'G_12': ['G_3', 'G_30'],
    'G_13': ['G_3', 'G_31'],
    'G_14': ['G_4', 'G_32'],
    'G_15': ['G_4', 'G_33'],
    'G_16': ['G_5', 'G_34'],
    'G_17': ['G_5', 'G_35'],
    'G_18': ['G_6', 'G_36'],
    'G_19': ['G_6', 'G_37'],
    'G_20': ['G_7', 'G_38'],
    'G_21': ['G_7', 'G_39'],
    'G_22': ['G_8', 'G_40'],
    'G_23': ['G_8', 'G_41'],
    'G_24': ['G_9', 'G_42'],
    'G_25': ['G_9', 'G_43'],
    'G_26': ['G_10', 'G_44'],
    'G_27': ['G_10', 'G_45'],
    'G_28': ['G_11', 'G_46'],
    'G_29': ['G_11', 'G_47'],
    'G_30': ['G_12', 'G_40', 'G_48'],
    'G_31': ['G_13', 'G_41', 'G_49'],
    'G_32': ['G_14', 'G_42', 'G_50'],
    'G_33': ['G_15', 'G_43'],
    'G_34': ['G_16', 'G_44'],
    'G_35': ['G_17', 'G_45'],
    'G_36': ['G_18', 'G_46'],
    'G_37': ['G_19', 'G_47'],
    'G_38': ['G_20', 'G_48'],
    'G_39': ['G_21', 'G_49'],
    'G_40': ['G_22', 'G_30'],
    'G_41': ['G_23', 'G_31'],
    'G_42': ['G_24', 'G_32'],
    'G_43': ['G_25', 'G_33'],
    'G_44': ['G_26', 'G_34'],
    'G_45': ['G_27', 'G_35'],
    'G_46': ['G_28', 'G_36'],
    'G_47': ['G_29', 'G_37'],
    'G_48': ['G_30', 'G_38'],
    'G_49': ['G_31', 'G_39'],
    'G_50': ['G_32']
}

start = 'G_F'
max_depth = 5

infection_stage = defaultdict(list)
distance = {}  # Í∞Å ÎÖ∏ÎìúÍπåÏßÄÏùò Í±∞Î¶¨ Ï†ÄÏû•Ïö© ÎîïÏÖîÎÑàÎ¶¨
visited = {start}
queue = deque([(start, 0)])
distance[start] = 0  # Í∞êÏóºÏûêÎäî Í±∞Î¶¨ 0

while queue:
    node, depth = queue.popleft()
    if depth > max_depth:
        continue
    infection_stage[depth].append(node)
    for neighbor in graph_50.get(node, []):
        if neighbor not in visited:
            visited.add(neighbor)
            distance[neighbor] = depth + 1  # Í±∞Î¶¨ Ï†ÄÏû•
            queue.append((neighbor, depth + 1))

print("üìä [Í∞êÏóº ÏãúÎÆ¨Î†àÏù¥ÏÖò Í≤∞Í≥º]")
print(f"Í∞êÏóºÏûê: {start}")
print("=" * 50)

for stage in range(1, max(infection_stage.keys()) + 1):
    members = infection_stage[stage]
    # Í∞Å Í∞êÏóºÏûêÏùò Í±∞Î¶¨ Ï†ïÎ≥¥ Ìï®Íªò Ï∂úÎ†•
    detailed = [f"{name}(Í±∞Î¶¨: {distance[name]})" for name in members]
    print(f"‚ñ∂ {stage}Îã®Í≥Ñ Í∞êÏóº ({len(members)}Î™Ö): {', '.join(detailed)}")
```
